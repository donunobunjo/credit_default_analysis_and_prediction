```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("credit.csv")

# Distribution of 'amount'
plt.figure(figsize=(8,5))
sns.histplot(df['amount'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Credit Amount')
plt.xlabel('Credit Amount')
plt.ylabel('Frequency')
plt.show()
```

🔍 Interpretation:

When you run this:

You’ll likely notice that the distribution is right-skewed, meaning most applicants request smaller credit amounts, but a few request very large amounts.

The kde (kernel density estimate) curve helps visualize the probability distribution more smoothly.

```python
#  Credit History Analysis

plt.figure(figsize=(8,5))
sns.countplot(x='credit_history', data=df, palette='viridis')

plt.title('Distribution of Credit History Categories')
plt.xlabel('Credit History')
plt.ylabel('Number of Applicants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

🔍 Interpretation:

Each bar shows how many applicants fall into each credit history category

```python
# Group by credit_history and default, then calculate proportions
proportion = (
    df.groupby(['credit_history', 'default'])
    .size()
    .groupby(level=0)
    .apply(lambda x: x / x.sum())
    .unstack()
)

print(proportion)
```

🔍 Explanation:

This table shows the proportion (share) of bad and good credits within each credit history category.

For example perfect has the highest proportion of defaulters while critical has the least proportion of defaulters

```python
## Relationship between credit amount and duration

plt.figure(figsize=(8,5))
sns.scatterplot(x='months_loan_duration', y='amount', data=df, color='teal', alpha=0.6)

plt.title('Relationship between Credit Amount and Duration')
plt.xlabel('Duration (months)')
plt.ylabel('Credit Amount')
plt.tight_layout()
plt.show()
```

🔍 Interpretation:

Each point represents a credit applicant.

You’ll likely notice an upward trend — as duration increases, credit amount also tends to increase.

```python
# Calculate correlation between credit amount and duration
corr = df['amount'].corr(df['months_loan_duration'])
print(f"Correlation between Credit Amount and Duration: {corr:.2f}")
```

🔍 Interpretation:

Shows a positive correlation 0.62, meaning higher credit amounts are associated with longer repayment periods.

```python
# Credit purpose analysis
plt.figure(figsize=(9,5))
sns.countplot(x='purpose', data=df, palette='viridis')

plt.title('Distribution of Credit Purposes')
plt.xlabel('Credit Purpose')
plt.ylabel('Number of Applicants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

🔍 Interpretation:

Each bar represents how many applicants took credit for each purpose (e.g., car, furniture/equipment, education, business, renovationss, etc.).

This helps identify the most common loan purposes — for example, must people took loans for furniture/appliances and least people took loan for used cars.

It also gives insight into the economic behavior of borrowers.

```python
# Group by purpose and default, then calculate proportions
proportion = (
    df.groupby(['purpose', 'default'])
    .size()
    .groupby(level=0)
    .apply(lambda x: x / x.sum())
    .unstack()
)

print(proportion)
```

🔍 Explanation:

This table shows the proportion (share) of bad and good credits within each loan purpose.

For example, credit for fairly used car has the highest proportion of defaulters
