# 📌 Credit Default Analysis and Prediction  

## 📖 Project Overview  
This project analyzes and predicts **loan default risk** using the Credit Dataset.  

**Goals:**  
- 🔎 Identify key financial and demographic risk factors for credit default  
- 🤖 Build predictive models to classify applicants as *low risk* or *high risk*  
- 📊 Provide insights to support better lending decisions  

---

## 🎯 Objectives  
1. Perform **Exploratory Data Analysis (EDA)**  
2. Clean & preprocess the dataset (missing values, duplicates, encoding)  
3. Engineer features (loan-to-income ratio, age groups, etc.)  
4. Train and evaluate multiple machine learning models  
5. Visualize and interpret the results  

---

## 🛠️ Tech Stack  
- 🐍 Python  
- 📊 Pandas / NumPy → Data wrangling  
- 🎨 Matplotlib / Seaborn → Visualization  
- 🤖 Scikit-learn → ML models & metrics  
- 📓 Jupyter Notebook → Analysis & reporting  

---

## 📊 Dataset  
- **Name:** Credit Dataset  
- **Source:** [UCI Machine Learning Repository](https://archive.ics.uci.edu/dataset/144/statlog+german+credit+data)  


## 🔍 Exploratory Data Analysis (EDA)  
Questions explored:  
- 👤 Does **age** influence default risk?  
- 💰 Are low **savings/checking balances** linked to higher default?  
- 🏠 Does **housing status** (own, rent, free) affect repayment?  
- 📈 How does **loan amount vs. income** affect creditworthiness?  

📌 Visualizations:  
- Distribution plots (age, loan amount, duration)  
- Default rate by housing, employment, credit history  
- Correlation heatmap

---

## 🤖 Machine Learning Approach  
**Models tested:**  
- Logistic Regression (baseline)  
- Decision Tree  
- Random Forest

**Metrics used:**  
- ✅ Accuracy  
- 🎯 Precision, Recall, F1-score  
- 📈 AUC-ROC Curve  

---

## ✅ Results & Insights

- Housing Status : Owning a house correlates with better credit performane.

- Purpose : Those applicants whose credit purposes are for Furniture/Appliances and new cars show lower risk. While those for used cars and education show very high rkis.

- Employement duration : 
Having stable job generally correlates with better credit performance.

- Savings Balance : Low savings balance and unknown savings balance seems to show the lowest default rates

- Age Matters : Young applicants (within the ages of 25 and 40) are associated with the highest default rates, while applicants less than 25 years show lower risk

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import (OneHotEncoder, StandardScaler, LabelBinarizer)
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (classification_report, confusion_matrix, accuracy_score, roc_auc_score, roc_curve)
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load dataset
df = pd.read_csv("credit.csv")
```

## Exploratory Data Analysis(EDA)

```python
# Preview data
print(df.head())
print(df.info())
print(df.describe(include='all'))
```

```python
# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Drop duplicates if any
if duplicates > 0:
    df = df.drop_duplicates()
```

```python
# 2. Check for missing values
missing = df.isnull().sum()
print("Missing values:\n", missing)
```

```python
# Fill missing values for categorical values
df['checking_balance'] = df['checking_balance'].fillna('unknown')
df['savings_balance'] = df['savings_balance'].fillna('unknown')
df['employment_duration'] = df['employment_duration'].fillna('unknown')
```

```python
# Numerical columns: fill with median
for col in ['months_loan_duration','amount','percent_of_income','years_at_residence','age','existing_loans_count','dependents']:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Basic statistics
print(df.describe())
```

```python
# detailed visualizations functions
def create_detailed_visualizations(df):
   
    # Set up the plotting area
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('Detailed Loan Data Analysis', fontsize=16)
    
    # 1. Loan amount distribution
    if 'amount' in df.columns:
        axes[0, 0].hist(df['amount'], bins=30, alpha=0.7, color='blue')
        axes[0, 0].set_title('Loan Amount Distribution')
        axes[0, 0].set_xlabel('Loan Amount')
        axes[0, 0].set_ylabel('Frequency')
    
    # 2. Loan duration distribution
    if 'months_loan_duration' in df.columns:
        axes[0, 1].hist(df['months_loan_duration'], bins=30, alpha=0.7, color='green')
        axes[0, 1].set_title('Loan Duration Distribution')
        axes[0, 1].set_xlabel('Months')
        axes[0, 1].set_ylabel('Frequency')
    
    # 3. Age distribution
    if 'age' in df.columns:
        axes[0, 2].hist(df['age'], bins=30, alpha=0.7, color='orange')
        axes[0, 2].set_title('Age Distribution')
        axes[0, 2].set_xlabel('Age')
        axes[0, 2].set_ylabel('Frequency')
    
    # 4. Purpose of loan
    if 'purpose' in df.columns:
        purpose_counts = df['purpose'].value_counts().head(10)
        purpose_counts.plot(kind='bar', ax=axes[1, 0], color='purple')
        axes[1, 0].set_title('credit purpose distribution')
        axes[1, 0].set_xlabel('Purpose')
        axes[1, 0].set_ylabel('Count')
        plt.sca(axes[1, 0])
        plt.xticks(rotation=45)
    
    # 5. Credit history
    if 'credit_history' in df.columns:
        credit_counts = df['credit_history'].value_counts()
        credit_counts.plot(kind='bar', ax=axes[1, 1], color='red')
        axes[1, 1].set_title('Credit History Distribution')
        axes[1, 1].set_xlabel('Credit History')
        axes[1, 1].set_ylabel('Count')
        plt.sca(axes[1, 1])
        plt.xticks(rotation=45)
    
    # 6. Employment duration
    if 'employment_duration' in df.columns:
        employment_counts = df['employment_duration'].value_counts()
        employment_counts.plot(kind='bar', ax=axes[1, 2], color='brown')
        axes[1, 2].set_title('Employment Duration')
        axes[1, 2].set_xlabel('Employment Duration')
        axes[1, 2].set_ylabel('Count')
        plt.sca(axes[1, 2])
        plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
```

```python
create_detailed_visualizations(df)
```

```python
# Define target and features
# -----------------------------
y = df["default"]
X = df.drop("default", axis=1)
```

```python
# Separate numeric and categorical features
numeric_features = ["months_loan_duration", "amount", "percent_of_income", "years_at_residence", "age", "existing_loans_count", "dependents"]

categorical_features = ["checking_balance", "credit_history", "purpose", "savings_balance", "employment_duration","other_credit", "housing", "job", "phone"]
```

```python
# Preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(drop="first"), categorical_features)
    ]
)
```

```python
# Model Pipelines
log_reg_pipe = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", LogisticRegression(max_iter=1000, random_state=42))
])

tree_pipe = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", DecisionTreeClassifier(random_state=42))
])

forest_pipe = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(random_state=42))
])
```

```python
# Parameter Grids
param_grid_log_reg = {
    "classifier__C": [0.01, 0.1, 1, 10],
    "classifier__solver": ["lbfgs", "liblinear"]
}

param_grid_tree = {
    "classifier__max_depth": [3, 5, 7, 10, None],
    "classifier__min_samples_split": [2, 5, 10],
    "classifier__criterion": ["gini", "entropy"]
}

param_grid_forest = {
    "classifier__n_estimators": [50, 100, 200],
    "classifier__max_depth": [5, 8, 10, None],
    "classifier__min_samples_split": [2, 5, 10],
    "classifier__criterion": ["gini", "entropy"]
}
```
