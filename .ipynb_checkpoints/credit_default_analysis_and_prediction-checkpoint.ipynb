````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
````

````python
# Load dataset
df = pd.read_csv("credit.csv")
````

## Exploratory Data Analysis(EDA)

````python
# Preview data
print(df.head())
print(df.info())
print(df.describe(include='all'))
````

````python
# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
````

````python
# Drop duplicates if any
if duplicates > 0:
    df = df.drop_duplicates()
````

````python
# 2. Check for missing values
missing = df.isnull().sum()
print("Missing values:\n", missing)
````

````python
# Fill missing values for categorical values
df['checking_balance'] = df['checking_balance'].fillna('unknown')
df['savings_balance'] = df['savings_balance'].fillna('unknown')
df['employment_duration'] = df['employment_duration'].fillna('unknown')
````

````python
# Numerical columns: fill with median
for col in ['months_loan_duration','amount','percent_of_income','years_at_residence','age','existing_loans_count','dependents']:
    df[col] = df[col].fillna(df[col].median())
````

````python
# Basic statistics
print(df.describe())
````

## Visualizations

````python
# Distribution of target variable (default)
sns.countplot(x='default', data=df)
plt.title("Default Distribution")
plt.show()
````

````python
# Numerical features distributions
num_cols = ['months_loan_duration','amount','percent_of_income','years_at_residence','age','existing_loans_count','dependents']
df[num_cols].hist(figsize=(12,8), bins=20)
plt.suptitle("Numerical Features Distribution")
plt.show()
````

````python
# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
````

````python
# Categorical feature analysis vs default
cat_cols = ['checking_balance','credit_history','purpose','savings_balance','employment_duration','other_credit','housing','job','phone']
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, hue='default', data=df)
    plt.xticks(rotation=45)
    plt.title(f"{col} vs Default")
    plt.show()
````

````python
# Outlier detection-
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='default', y=col, data=df)
    plt.title(f"Boxplot of {col} by Default")
    plt.show()
````

````python
# cleaned_df = perform_eda(df)

# If you want to see specific visualizations for your features:
def create_detailed_visualizations(df):
    """
    Create more detailed visualizations for the loan dataset
    """
    # Set up the plotting area
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('Detailed Loan Data Analysis', fontsize=16)
    
    # 1. Loan amount distribution
    if 'amount' in df.columns:
        axes[0, 0].hist(df['amount'], bins=30, alpha=0.7, color='blue')
        axes[0, 0].set_title('Loan Amount Distribution')
        axes[0, 0].set_xlabel('Loan Amount')
        axes[0, 0].set_ylabel('Frequency')
    
    # 2. Loan duration distribution
    if 'months_loan_duration' in df.columns:
        axes[0, 1].hist(df['months_loan_duration'], bins=30, alpha=0.7, color='green')
        axes[0, 1].set_title('Loan Duration Distribution')
        axes[0, 1].set_xlabel('Months')
        axes[0, 1].set_ylabel('Frequency')
    
    # 3. Age distribution
    if 'age' in df.columns:
        axes[0, 2].hist(df['age'], bins=30, alpha=0.7, color='orange')
        axes[0, 2].set_title('Age Distribution')
        axes[0, 2].set_xlabel('Age')
        axes[0, 2].set_ylabel('Frequency')
    
    # 4. Purpose of loan
    if 'purpose' in df.columns:
        purpose_counts = df['purpose'].value_counts().head(10)
        purpose_counts.plot(kind='bar', ax=axes[1, 0], color='purple')
        axes[1, 0].set_title('Top 10 Loan Purposes')
        axes[1, 0].set_xlabel('Purpose')
        axes[1, 0].set_ylabel('Count')
        plt.sca(axes[1, 0])
        plt.xticks(rotation=45)
    
    # 5. Credit history
    if 'credit_history' in df.columns:
        credit_counts = df['credit_history'].value_counts()
        credit_counts.plot(kind='bar', ax=axes[1, 1], color='red')
        axes[1, 1].set_title('Credit History Distribution')
        axes[1, 1].set_xlabel('Credit History')
        axes[1, 1].set_ylabel('Count')
        plt.sca(axes[1, 1])
        plt.xticks(rotation=45)
    
    # 6. Employment duration
    if 'employment_duration' in df.columns:
        employment_counts = df['employment_duration'].value_counts()
        employment_counts.plot(kind='bar', ax=axes[1, 2], color='brown')
        axes[1, 2].set_title('Employment Duration')
        axes[1, 2].set_xlabel('Employment Duration')
        axes[1, 2].set_ylabel('Count')
        plt.sca(axes[1, 2])
        plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
````

````python
create_detailed_visualizations(df)
````
